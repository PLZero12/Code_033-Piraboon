```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

digits = load_digits()
X, y = digits.data, digits.target

tsne = TSNE(n_components=2, random_state=42)
X_tane = tsne.fit_transform(X)

plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_tane[:, 0], X_tane[:, 1], c=y, cmap='viridis')
plt.colorbar(scatter)
plt.xlabel('t-SNE feature 1')
plt.ylabel("t-SNE feature 2")
plt.title('t-SNE cisualization of MNIST digits')
plt.show()

print("Origina; data shape:", X.shape)
print("t-SNE reduced data shape:", X_tane.shape)
```
